```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize
```

```python
def fit_func(_x, _a, _b):
    # budget = a * timestamp + b
    return _x * _a + _b
```

```python
x = [1716481144, 1716495526, 1716509933, 1716523050, 1716524813]
y = [84.21, 80.16, 70.46, 48.71, 44.83]
```

```python
plt.figure()
plt.scatter(x, y, 25, "red")

param_a, param_b = optimize.curve_fit(fit_func, x, y)[0]
param_c, param_d = optimize.curve_fit(fit_func, x[:3], y[:3])[0]

x_p = range(x[0], x[-1], 100)
y_p = param_a * x_p + param_b
y_q = param_c * x_p + param_d

plt.plot(x_p, y_p, "blue")
plt.plot(x_p, y_q, "green")
plt.show()
```
