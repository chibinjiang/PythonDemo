```python
import requests
import pandas as pd
```

```python
df = pd.read_csv('./openfood_100000_products.tsv', sep='\t')
df
```

```python
type_mapping = {
    'float': 'Decimal64(6)', 
    'str': 'String',
    'int': 'Int64'
}
type_default = {
    'float': 0, 
    'str': "''",
    'int': 0
}

for k, v in sorted(df.iloc[0].to_dict().items(), key=lambda x: x[0]):
    k = k.replace('-', '_')
    t_name = type(v).__name__
    default_v = type_default[t_name]
    # print(k, "\t", t_name)
    print(f"{k}\t\t{type_mapping[t_name]}\tDEFAULT {default_v}\tCOMMENT '{k}',")
```

```python
openfood_resp = requests.get("https://world.openfoodfacts.org/api/v0/product/3229820019307.json")
api_df = pd.DataFrame([openfood_resp.json()['product']])
api_df
```

```python
for k, v in api_df.iloc[0].to_dict().items():
    print("Key: ", k, "\tValue: ", v, "\tType: ", type(v))
```
