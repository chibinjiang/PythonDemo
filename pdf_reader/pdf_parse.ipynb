```python
import os

# 紫苏黄瓜炒肉

def validate_pdf(folder):
    for file in os.listdir(folder):
        if not file.endswith('pdf'):
            continue
        with open(file, 'rb') as file_obj:
            print(f"Validate: {file}")
            _pdf_reader = PyPDF2.PdfReader(file_obj)
            page_count = len(_pdf_reader.pages)
            first_page = _pdf_reader.pages[0]
            print(first_page.extract_text())    
            
    
validate_pdf("./")
```
